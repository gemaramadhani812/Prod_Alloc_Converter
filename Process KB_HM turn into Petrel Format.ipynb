```python
#Import Packages
import pandas as pd
from pathlib import Path
```

```python
#Import Data yang akan di convert to Petrel File
df = pd.read_csv("data/KB_HM.csv", parse_dates=["DATE"])
df.head()
```

```python
#Data Processing (Mengeluarkan DATE)
df["MONTH"] = df.DATE.dt.month
df["YEAR"] = df.DATE.dt.year
df.drop(columns = (['DATE', 'CUMOILP', 'CUMWATP', 'CUMWATI', 'CUMLIQP', 'WC(%)']), inplace = True)
df.head()
```

```python
#Rename Columns jika perlu
df.rename(columns={
    'OPR_(stb/d)': "OIL",
    'WPR_(stb/d)': "WATER",
    'WINJ_(stb/d)': "WINJ",
    'LPR_(stb/d)': "LPR",
    'CUMGASP': "GAS",
}, inplace=True)
df.head()
```

```python
#Drop Columns yang tidak perlu
df.drop(columns = (['LPR']), inplace = True)
df.head()
```

### Process data turn to Petrel Format

```python
#Silahkan ganti nama file / folder output yang ingin dikeluarkan
outname = Path("output/KB_HM_2016.txt")
#Silahkan ganti nama columns dan urutannya
cols = ["DAY", "MONTH", "YEAR", "OIL", "GAS", "WATER", "WINJ"]
```

```python
df.WELLNAME
```

```python
outname.parent.mkdir(exist_ok=True)
with open(outname, "w") as f:
    # Tulis baris pertama
    f.write("\t".join([f"*{col}" for col in cols]) + "\n" + "\n")
    
    # Tulis data tiap sumur    
    for well in df.WELLNAME.unique():
        f.write("\n" + f"*WELLNAME {well}" + "\n")
        f.write(df[df.WELLNAME == well][cols].to_string(header=False, index=False) + "\n")
```
