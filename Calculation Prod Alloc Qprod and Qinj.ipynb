## 1. Introduction
Get to know to Prod Alloc Kotabatak:

    Pada sheet ini melakukan converting "Decimal Date" to "dd-mm-yyyy" dan perhitungan rerata Qprod dan Qinj (stb/d) setiap bulan. Berikut datanya:
    
    WELLNAME: Nama well (UWI Format);
    
    DATE: Decimal Date (1980.56);
    
    CUMOPIL: cumulative oil production, STB/Month;
    
    CUMWATP: cumulative water production, STB/Month;

    CUMWATI: cumulative water injection, STB/Month;

    CUMLIQP: cumulative liquid production, STB/Month;

```python
#Import Packages
import pandas as pd
from math import floor, ceil
from datetime import datetime, timedelta
```

```python
#Define decimal year to DD-MM-YY (Include leap year)
def decimal_year_to_datetime_with_leap_year(t):
    year_approx = datetime(ceil(t), 1, 1) - datetime(floor(t), 1, 1)
    dt = datetime(floor(t), 1, 1) + timedelta((t - floor(t))*year_approx.days)
    return dt
```

```python
#Import decimal date data
df = pd.read_csv("data/KB_HM_2016.csv", nrows=226950)

#Define the data to convert
df.DATE = df.DATE.apply(decimal_year_to_datetime_with_leap_year)
df.DATE = df.set_index("DATE").to_period("M").to_timestamp('M').reset_index().DATE.dt.strftime("%d-%b-%y")

#Import new data (2016)
df2 = pd.read_csv("data/KB_HM_2016.csv", skiprows=226950, names=df.columns)
df = df.append(df2).reset_index()
```

```python
#Make a new columns
df['DAY'] = df.DATE
#And define based on calender date
df.DAY = pd.to_datetime(df.DAY)
df['DAY'] = df.DAY.dt.day
```

```python
#Calculate Prod Rate and Inj Rate
df['CUMOILP(stb/d)'] = df.CUMOILP/df.DAY
df['CUMWATP(stb/d)'] = df.CUMWATP/df.DAY
df['CUMWATI(stb/d)'] = df.CUMWATI/df.DAY
df['CUMLIQP(stb/d)'] = df.CUMLIQP/df.DAY
```
